```python
from Bio import Entrez, SeqIO
```

```python
import os
```

```python
Entrez.email = "sharmabhavya609@gmail.com"
search_term = "Cytochrome P450"
max_sequences = 200
```

```python
output_path = "../data/cytochrome_p450_sequences.fasta"
```

```python
print("Searching NCBI for Cytochrome P450 protein sequences")
```

```python
handle = Entrez.esearch(
    db="protein",
    term=search_term,
    retmax=max_sequences
)
record = Entrez.read(handle)
handle.close()
```

```python
ids = record["IdList"]
print(f"Found {len(ids)} sequences.")
```

```python
print("Downloading all sequences")
```

```python
fetch_handle = Entrez.efetch(
    db="protein",
    id=",".join(ids),
    rettype="fasta",
    retmode="text"
)
```

```python
fasta_data = fetch_handle.read()
fetch_handle.close()
```

```python
with open(output_path, "w") as f:
    f.write(fasta_data)
```

```python
print(f"\nSaved {len(ids)} sequences to: {output_path}")
```
