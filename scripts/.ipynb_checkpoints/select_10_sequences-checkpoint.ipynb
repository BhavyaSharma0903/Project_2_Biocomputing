```python
from Bio import SeqIO
import os
```

```python
input_fasta = "../data/cytochrome_p450_sequences_filtered.fasta"
output_dir = "../data/ten_sequences/"
```

```python
records = list(SeqIO.parse(input_fasta, "fasta"))
selected = {}
```

```python
for record in records:
    description = record.description
    if "[" in description and "]" in description:
        organism = description.split("[")[-1].split("]")[0].strip()
        if organism not in selected:
            selected[organism] = record
        if len(selected) == 10:
            break
```

```python
for organism, rec in selected.items():
    safe_name = organism.replace(" ", "_").replace("/", "_")
    output_path = f"{output_dir}{safe_name}.fasta"
    SeqIO.write(rec, output_path, "fasta")
    print("Saved:", output_path)
```

```python
print("\nTotal sequences saved:", len(selected))
```
