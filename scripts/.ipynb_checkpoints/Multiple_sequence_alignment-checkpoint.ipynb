```python
from Bio.Align.Applications import ClustalOmegaCommandline
from Bio import AlignIO, Phylo
```

```python
fasta_in = "../data/combined_10_sequences.fasta"
align_out = "sequences_aligned.fasta"
tree_out = "sequences_tree.dnd"
```

```python
clustalo_cline = ClustalOmegaCommandline(
    infile=fasta_in,
    outfile=align_out,
    guidetree_out=tree_out,
    force=True,
    outfmt="fasta"
)
```

```python
print("Running Clustal Omega...")
stdout, stderr = clustalo_cline()
print("Done.")
```

```python
alignment = AlignIO.read(align_out, "fasta")
print(f"Number of sequences: {len(alignment)}")
print(f"Alignment length: {alignment.get_alignment_length()}")
```

```python
print(alignment)
```

```python
print("\nReading tree:\n")
tree = Phylo.read(tree_out, "newick")
Phylo.draw_ascii(tree)
```
