```python
import matplotlib
matplotlib.use("Agg")    

import matplotlib.pyplot as plt
from Bio import SeqIO
import os
```

```python
fasta_file = "../data/cytochrome_p450_sequences_filtered.fasta"
```

```python
lengths = [len(record.seq) for record in SeqIO.parse(fasta_file, "fasta")]

print(f"Total sequences loaded: {len(lengths)}")
print(f"Min length: {min(lengths)}, Max length: {max(lengths)}")
```

```python
plt.figure(figsize=(10,6))
plt.hist(lengths, bins=20)
plt.xlabel("Sequence Length (amino acids)")
plt.ylabel("Count")
plt.title("Distribution of Sequence Lengths (Cytochrome P450)")
plt.grid(True)
```

```python
output_path = "../results/sequence_length_distribution.jpg"
plt.savefig(output_path, dpi=300, format='jpg', bbox_inches='tight')
plt.close()
print(f"Plot saved successfully to: {output_path}")
```
